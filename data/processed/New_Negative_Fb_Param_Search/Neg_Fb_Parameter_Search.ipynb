```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import sys, os
import numpy as np
sys.path.append("/mit/rganti/SSC_python_modules/")
import matplotlib.pyplot as plt
from output_information import HistogramInformation, OneDHistogramInformation
```

```python
hist_info = HistogramInformation(directory="../New_Negative_Fb_Param_Search", lf=30)
```

```python
hist_info.file_paths
```

```python
hist_info.compute_variables(rate='k_neg_fb')
```

```python
hist_info.ic
```

```python
hist_info.rate_variable
```

```python
plt.imshow(np.array(hist_info.ic), extent=[hist_info.rate_variable[0][0], hist_info.rate_variable[0][-1], 5.5, 0.5], 
           aspect='auto', vmin=0.0, vmax=1.0)
plt.colorbar(orientation='vertical')
x_labels = [str(i) for i in hist_info.rate_variable[0]]
on_rates = [0.0, 0.5, 1.0, 2.0, 3.0, 5.0]
y_labels = [str(i) for i in on_rates]
plt.locator_params(axis='y', nbins=6)
plt.locator_params(axis='x', nbins=4)

ax = plt.gca()
ax.set_xticklabels(x_labels, ha='center')
ax.set_yticklabels(y_labels)

plt.ylabel("$k_{on}$", size=14)
plt.xlabel("$k_{neg,fb}$", size=14)
plt.title("$L_{f} = 30$", size=20)
# plt.savefig("../New_Negative_Fb_Param_Search/neg_fb_lf_30.pdf", format='pdf')
```
