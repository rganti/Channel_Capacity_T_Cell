```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import sys, os
import numpy as np
sys.path.append("/mit/rganti/SSC_python_modules/")
import matplotlib.pyplot as plt
from compute_ic import InformationCapacity
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from matplotlib.ticker import FormatStrFormatter

```

```python
num_steps = []
capacity = []

for i in range(1, 9):
    if i != 5:
        file_path = "{0}_step/".format(i)
        ic_lf = InformationCapacity(foreign_directory=file_path + "Ls_Lf_{0}/".format(30),
                                    self_directory=file_path + "Ls/", limiting="self")
        num_steps.append(i)
        capacity.append(ic_lf.capacity)
```

```python
num_steps_1 = []
capacity_1 = []

for i in range(1, 9):
    if i != 5:
#         if i == 8:
#             file_path = "k_on_1.0_latpp/8_step_slow_latpp/".format(i)
#         else:
        file_path = "../../Standard_Model/k_on_1.0_latpp/{0}_step/".format(i)
        ic_lf = InformationCapacity(foreign_directory=file_path + "Ls_Lf_{0}/".format(30),
                                    self_directory=file_path + "Ls/", limiting="self")
        num_steps_1.append(i)
        capacity_1.append(ic_lf.calculate_ic())

slow_num_steps_1 = []
slow_capacity_1 = []

for i in range(7, 9):
    if i == 8:
        file_path = "../../Standard_Model/k_on_1.0_latpp/8_step_slow_latpp/".format(i)
    else:
        file_path = "../../Standard_Model/k_on_1.0_latpp/{0}_step/".format(i)
    ic_lf = InformationCapacity(foreign_directory=file_path + "Ls_Lf_{0}/".format(30),
                                self_directory=file_path + "Ls/", limiting="self")
    slow_num_steps_1.append(i)
    slow_capacity_1.append(ic_lf.calculate_ic())
```

```python
plt.rcParams["figure.figsize"] = (8,6)
```

```python
plt.plot(range(1,8), capacity_1, linestyle='-', marker='o', color='blue', 
         label="$k_{on} = 1.0 s^{-1}, k_{Y132} = 1.0 s^{-1}$")
plt.plot(range(6,8), slow_capacity_1, linestyle='-', marker='o', color='orange',
         label="$k_{on} = 1.0 s^{-1}, k_{Y132} = 0.1 s^{-1}$")


plt.plot(range(1,8), capacity, linestyle='-', marker='o', color='green',
         label="$k_{on} = 2.0 s^{-1}, k_{Y132} = 2.0 s^{-1}$")
plt.plot(range(6,8), slow_capacity, linestyle='-', marker='o', color='red', 
         label="$k_{on} = 2.0 s^{-1}, k_{Y132} = 0.2 s^{-1}$")
# plt.plot(range(1,8), slow_early_capacity, linestyle='-', marker='o', color='grey',
#          label="$k_{on} = 2.0 s^{-1}, k_{rpl\_lck} = 0.2 s^{-1}$")

plt.xlabel("Number of Steps", size=20)
plt.ylabel("C (bits)", size=20)
plt.legend(loc=4, frameon = False, prop={"size":12})
plt.xlim(0, 7)
plt.ylim(0, 1)

ax = plt.gca()
axins = ax.inset_axes([0.098, 0.58, 0.33, 0.33])
axins.plot(range(1,8), capacity_1, ls = '-', marker='o', color='blue', 
           label='$\sigma(\\tau)$')
axins.plot(range(6,8), slow_capacity_1, ls = '-', marker='o', 
           label='$\sigma_{\\theta}(\\tau)$', color='orange')
axins.set_ylim(0.9, 1.0)
axins.set_xlim(5, 7)
axins.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))

plt.savefig("C_vs_8_Fig_2.pdf", format="pdf")
# plt.savefig("C_vs_8_bound_latpp_early_late.pdf", format="pdf")
# plt.savefig("C_vs_8_bound_latpp_Y132.pdf", format="pdf")
# plt.savefig("C_vs_8_bound_latpp.pdf", format="pdf")
```
